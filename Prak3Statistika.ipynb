```python
import pandas as pd

data_dewanto = pd.read_clipboard()

# Menampilkan data
print(data_dewanto)
```

```python
dewanto = data_dewanto[data_dewanto['Bedrooms'] == 2]

# Menampilkan nama
print(dewanto)
```

```python
dewanto['Bathrooms'] = pd.to_numeric(dewanto['Bathrooms'])

import numpy as np

dewanto['Bathrooms'] = dewanto['Bathrooms'].apply(lambda x: 'large' if x > 2 else 'small')

# Menampilkan DataFrame setelah modifikasi
print(dewanto)
```

```python
import numpy as np

dewanto['newvariable'] = np.where(dewanto['Offers'] > 2, 'large', 'small')

# Menampilkan DataFrame 'nama' setelah penambahan kolom baru
print(dewanto)
```

```python
# Menambahkan kolom baru 'newvariable' 
dewanto['newvariable'] = dewanto['Price'] / dewanto['SqFt']

# Menampilkan DataFrame 'nama' setelah penambahan kolom baru
print(dewanto)
```

```python
dewanto = dewanto.drop(columns=['newvariable'])

# Menampilkan DataFrame 'nama' 
print(dewanto)
```

```python
kolom1dan2 = data_dewanto.iloc[:, 0:2]

print(kolom1dan2)
```

```python
# Memilih kolom 1 dan 2 dari DataFrame data_nama
kolom3dan4 = data_dewanto.iloc[:, 2:4]

# Menampilkan DataFrame kolom3dan4
print(kolom3dan4)
```

```python
kolom1sd4 = pd.concat([kolom1dan2, kolom3dan4], axis=1)

# Menampilkan DataFrame kolom1sd4
print(kolom1sd4)
```
